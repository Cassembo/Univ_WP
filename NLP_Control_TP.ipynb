```python
import pandas as pd

# Load the uploaded dataset to examine it
file_path = '/content/medium_data.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
data.info()
```

```python
# Extract the 'title' column into a separate DataFrame
titles_df = data[['title']]

# Display the first few rows of the new DataFrame
titles_df.head()
```

```python
# Clean the titles by removing non-alphanumeric symbols and renaming the DataFrame
titles_df['title'] = titles_df['title'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

# Rename the DataFrame
t_data = titles_df

# Display the first few cleaned titles
t_data.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK resources (only needs to be run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# List of English stop words
stop_words = set(stopwords.words('english'))

# Preprocess titles
def preprocess_title(title):
    title = title.lower()  # Convert to lowercase
    tokens = word_tokenize(title)  # Tokenize the title
    filtered_tokens = [word for word in tokens if word not in stop_words and word.isalnum()]  # Remove stop words and non-alphanumeric tokens
    return filtered_tokens

# Apply preprocessing to all titles
t_data['title'] = t_data['title'].apply(preprocess_title)

# Display the first few processed rows
print(t_data.head())
```

```python
t_data.head()
```

```python
from nltk.util import ngrams

# Function to generate n-grams
def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))

# Preprocess titles and generate n-grams
n = 2  # Example: bigram model
t_data['ngrams'] = t_data['title'].apply(lambda tokens: generate_ngrams(tokens, n))

# Display the first few rows with n-grams
t_data[['title', 'ngrams']].head()
```

```python
from collections import Counter
all_ngrams = t_data['ngrams'].explode()
ngram_counts = Counter(all_ngrams)
print(ngram_counts.most_common(10))
```

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

# Preprocessing function (assumes cleaning is already done in your data)
def preprocess_text(text):
    text = " ".join(text)
    return word_tokenize(text.lower())  # Lowercase and tokenize

# Generate incremental N-grams (from 1 to length of the text)
def generate_incremental_ngrams(tokens):
    max_n = len(tokens)
    all_ngrams = []
    for n in range(1, max_n + 1):  # From unigram to full sentence
        all_ngrams.extend(list(ngrams(tokens, n)))
    return all_ngrams

# Apply to all titles in the dataset
t_data['incremental_ngrams'] = t_data['title'].apply(lambda title: generate_incremental_ngrams(preprocess_text(title)))

# Display the first few results
t_data[['title', 'incremental_ngrams']].head()
```

```python
from nltk.tokenize import word_tokenize

# Preprocess titles (tokenization)
t_data['tokenized_title'] = t_data['title'].apply(lambda tokens: [word.lower() for word in tokens]) # Lowercase each token in the list

# Find the maximum length of tokens in the dataset
max_length = t_data['tokenized_title'].apply(len).max()

# Function to pad titles to the maximum length
def pad_title(tokens, max_length):
    return tokens + ['PAD'] * (max_length - len(tokens))

# Pad all titles to the maximum length
t_data['padded_title'] = t_data['tokenized_title'].apply(lambda tokens: pad_title(tokens, max_length))

# Display the first few padded titles
print(f"Maximum Length: {max_length}")
t_data.head()
```

```python
from collections import defaultdict

# Flatten the list of tokenized titles to create a vocabulary
all_tokens = [token for tokens in t_data['tokenized_title'] for token in tokens]

# Create a vocabulary mapping each word to a unique integer
vocab = {word: idx + 1 for idx, word in enumerate(set(all_tokens))}  # Start indexing from 1
vocab['PAD'] = 0  # Assign 0 for the padding token

# Function to tokenize titles numerically
def numerical_tokenize(tokens, vocab):
    return [vocab[token] for token in tokens]

# Apply numerical tokenization to the padded titles
t_data['numerical_title'] = t_data['padded_title'].apply(lambda tokens: numerical_tokenize(tokens, vocab))

# Display the first few results
print("Vocabulary Size:", len(vocab))
t_data[['tokenized_title', 'padded_title', 'numerical_title']].head()
```

```python
# Flatten the N-grams into a list of unique words to build the vocabulary
all_ngrams_tokens = [
    token for ngram_list in t_data['incremental_ngrams']
    for ngram in ngram_list
    for token in ngram
]

# Create vocabulary for N-grams (unique integers for each token)
vocab_ngrams = {word: idx + 1 for idx, word in enumerate(set(all_ngrams_tokens))}
vocab_ngrams['PAD'] = 0  # Assign 0 for the padding token

# Function to numerically tokenize N-grams
def numerical_tokenize_ngrams(ngrams_list, vocab):
    return [[vocab[token] for token in ngram] for ngram in ngrams_list]

# Apply numerical tokenization to incremental N-grams
t_data['numerical_ngrams'] = t_data['incremental_ngrams'].apply(lambda ngrams_list: numerical_tokenize_ngrams(ngrams_list, vocab_ngrams))

# Display results
print("Vocabulary Size for N-grams:", len(vocab_ngrams))
t_data[['incremental_ngrams', 'numerical_ngrams']].head()
```

```python
# Function to split data and label
def split_data_and_label(numerical_title):
    # Exclude padding (0) and split into data and label
    valid_tokens = [token for token in numerical_title if token != 0]
    if len(valid_tokens) < 2:  # Ensure there are enough tokens for data and label
        return None  # Mark invalid cases for filtering
    return {
        'data': valid_tokens[:-1],  # All except the last token
        'label': valid_tokens[-1]   # The last token
    }

# Apply the function to numerical_title and filter out invalid rows
t_data['data_label'] = t_data['numerical_title'].apply(split_data_and_label)
t_data = t_data[t_data['data_label'].notnull()]  # Remove rows where split_data_and_label returned None

# Extract data and label into separate columns for clarity
t_data['data'] = t_data['data_label'].apply(lambda x: x['data'])
t_data['label'] = t_data['data_label'].apply(lambda x: x['label'])

# Drop the intermediate column
t_data.drop(columns=['data_label'], inplace=True)

# Display results
t_data[['numerical_title', 'data', 'label']].head()
```
